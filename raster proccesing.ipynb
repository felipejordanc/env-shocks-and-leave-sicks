```python
import xarray as xr
import rioxarray  # this extends xarray with rasterio methods
import glob
import os

# Path to your folder
folder = r"C:\Users\black\temp daily"


# Get all .nc files
nc_files = sorted(glob.glob(os.path.join(folder, "*.nc")))

print(f"Found {len(nc_files)} NetCDF files.")

# Open and select variable 'tmin'
datasets = []
for f in nc_files:
    ds = xr.open_dataset(f)

    # Rename spatial dims if needed (adjust depending on your file!)
    if "lat" in ds.dims and "lon" in ds.dims:
        ds = ds.rename({"lat": "y", "lon": "x"})

    datasets.append(ds[["tmin"]])

# Concatenate along time
merged = xr.concat(datasets, dim="time")

# Register CRS (replace EPSG code with the correct one for your data)
merged_rio = merged["tmin"].rio.write_crs("EPSG:4326")

# Save to raster
output_file = os.path.join(folder, "tmin.tif")
merged_rio.rio.to_raster(output_file)

```
